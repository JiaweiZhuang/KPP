```python
%matplotlib inline
import warnings; warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('small_strato/small_strato.dat',
                delim_whitespace=True, header=None, index_col=0)
df.columns = ['SUN', 'O1D', 'O', 'O3', 'NO', 'NO2', 'M', 'O2']
df.index.name = 'time'

# convert "0.7861355928602839-314" to nan and then to 0.0
df = df.apply(pd.to_numeric, errors='coerce')
df[np.isnan(df)] = 0.0
```

```python
df.plot(y=['SUN', 'O1D', 'O', 'O3', 'NO', 'NO2'], #df.columns.drop('time'),
        ylim=[1e-2, None], logy=True, grid=True)
```
